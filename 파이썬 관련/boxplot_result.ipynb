```python
#!sudo apt-get install -y fonts-nanum
#!sudo fc-cache -fv
#!rm ~/.cache/matplotlib -rf

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

plt.rc('font', family='NanumBarunGothic')
result = pd.read_csv('/content/result_boxplot_개연성.csv')
idx = np.arange(1, 5)
labels = ['ChatGPT', 'Bard', 'ClovaX', 'Bing Chat']
means = [3.475, 3.05, 3.625, 2.9]

plt.figure(figsize = (10, 2))
plt.boxplot([result['chatGPT_개연성'], result['Bard_개연성'],
           result['ClovaX_개연성'], result['NewBing_개연성']])
plt.title('모델별 개연성')
plt.scatter(idx, means, color='red', marker='o', label='평균값')

#for i in range(len(idx)):
    #plt.text(idx[i], means[i], f'평균', fontsize=5, ha='right', va='bottom')

plt.xticks(idx, labels)
plt.show()
```
